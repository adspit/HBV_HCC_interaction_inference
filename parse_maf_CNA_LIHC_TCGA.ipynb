```python
# create dictionary with HepB status for each patient: 0 = negative, 1 = positive for HepB
# hepB_status.txt contains the name of each TCGA HCC sample and the Hepatitis B infection status 
# according to the TCGA HCC clinical data

barcode2hepB={}
for line in open('../hepB_status.txt').read().rstrip().split('\n'):
    row=line.split('\t')
    if row[1]=='TRUE':
        barcode2hepB[row[0][:12]]='1'
    elif row[1]=='FALSE':
        barcode2hepB[row[0][:12]]='0'
        
```

```python
# create dictionary with hepC status for each patient:  0 = negative, 1 = positive for HepC
# hepC_status.txt contains the name of each TCGA HCC sample and the Hepatitis C infection status 
# according to the TCGA HCC clinical data

barcode2hepC={}
for line in open('../hepC_status.txt').read().rstrip().split('\n'):
    row=line.split('\t')
    if row[1]=='TRUE':
        barcode2hepC[row[0][:12]]='1'
    elif row[1]=='FALSE':
        barcode2hepC[row[0][:12]]='0'
```

```python
# read in the oncogene file - a file that contains gene names and annotated information regarding status as oncogene,
# tumor suppressor 
oncogene_tsg={}
for line in open("oncogene_tsg.txt").read().rstrip().split("\n"):
    a=line.split("\t")
    oncogene_tsg[a[0]]=a[1]
```

```python
oncogene_tsg
```

```python
# function to parse through the mutations files
def parse_maf(fn, cohort):
    
    df = pd.read_table(fn, encoding = "ISO-8859-1", low_memory=False)
    
    df = df.loc[(df.loc[:,'is_flank']==0) & (df.loc[:,'is_silent']==0),:]
    if cohort[:4] == 'TCGA':
        df['pat'] = df.loc[:,'Tumor_Sample_Barcode'].str[:12]
        
    filter_rows = []
    for index, row in df.iterrows():
        gene = row['Hugo_Symbol']
        VC = row['Variant_Classification']
        if gene in oncogene_tsg and oncogene_tsg[gene]=='Oncogene':
            if VC not in ['Missense_Mutation', 'In_Frame_Del', 'In_Frame_Ins']:
                filter_rows.append(index)
    df = df.drop(filter_rows)
    
    df = df.loc[:,['pat','Hugo_Symbol']]
    
    df['counter']=1
    df.set_index(['pat','Hugo_Symbol'],inplace=True)
    
    df=df.counter.groupby(level=[0, 1]).min().unstack()
    df.fillna(0,inplace=True)
    
    return df
```

```python
# parse the TCGA LIHC mutation file - see TCGA data portal

fn='LIHC-TP.maf.annotated'
df = parse_maf(fn, 'TCGA-LIHC')
# df 198 rows (patients) Ã— 9782 genes
```

```python
# function to parse through the CNA file
def parse_CNA(fn, cohort):
    df = pd.read_table(fn,low_memory=False,index_col=0)
    if cohort[:4] == 'TCGA':
        df = df.iloc[:,2:]
        df = (df/2.).round(0)
        df.columns = df.columns.str[:12]
    nonOncogene_rows = []
    for index, row in df.iterrows():
        gene = index
        if not (gene in oncogene_tsg and oncogene_tsg[gene] in ['Oncogene', 'Amplification_Oncogene']):
            nonOncogene_rows.append(index)
    df.loc[nonOncogene_rows,:] = df.loc[nonOncogene_rows,:]*(-1)
    df = df.clip(lower=0)
    return df
```

```python
# parse the CNA LIHC file - Copy Number Gistic2 results file form TCGA data portal
fn = 'LIHC/CopyNumber_Gistic2/all_thresholded.by_genes.txt'
df4 = parse_CNA(fn, 'TCGA-LIHC')
# 24776 (genes/locations) Ã— 370 patients
```

```python
# concatenate CNA file with the mutation file
df = pd.concat([df,df4.transpose()],join='inner').groupby(level=0).sum().clip(upper=1.).loc[set(df.index) & set(df4.columns),:]
```

```python
# add two columns with the hepB, hepC status
hepB=[]
hepC=[]
for index, row in df.iterrows():
    if index in barcode2hepB and barcode2hepB[index]=='0':
        hepB.append(0)
    else:
        hepB.append(1)
    if index in barcode2hepC and barcode2hepC[index]=='0':
        hepC.append(0)
    else:
        hepC.append(1)        
df['hepB'] = hepB
df['hepC'] = hepC
df.to_csv('../data/df4glm_LIHC.txt',sep='\t')
# df, 193 patients Ã— 9294 genes
```
